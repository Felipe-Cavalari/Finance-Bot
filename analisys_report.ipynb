#### Importando libs

```python
import pandas as pd
from datetime import datetime
```

#### Criando as variaveis

```python
file_path = "./relatorios"
file_name = "altas_e_baixas"
date = datetime.now().strftime("%d_%m_%Y")
full_file = f"{file_path}/{file_name}_{date}.csv"

print(full_file)
```

### Analisa arquivo da data atual

```python
df = pd.read_csv(full_file)
display(df)
```

```python
# Vamos coletar as 5 maiores altas 
top5 = df.head(5)
top5
```

```python
# Vamos coletar as 5 maiores baixas
bottom5 = df.tail(5).sort_values(by='Var. Dia (%)', ascending=False)
bottom5
```
